# 📊 **3. Descripción de columnas del dataset***

A continuación se detalla el significado de cada una de las variables presentes en el dataset, así como su utilidad dentro del proyecto:

- **`Unnamed: 0`**  
  - Identificador artificial generado en un proceso previo de enriquecimiento.  
  - **Prescindible**, ya que cada POI dispone de un identificador propio.

- **`id`**  
  - Identificador principal y original de cada POI dentro del dataset.

- **`name`**  
  - Nombre del país asociado al POI.

- **`country_latitude`**  
  - Latitud del país.  
  - **Prescindible**, ya que no aporta valor a nivel de análisis local.

- **`country_longitude`**  
  - Longitud del país.  
  - **Prescindible** por la misma razón que la latitud del país.

- **`city`**  
  - Ciudad a la que pertenece el POI.  
  - **Prescindible**, dado que todos los registros corresponden a Barcelona.

- **`city_latitude`**  
  - Latitud de la ciudad.

- **`city_longitude`**  
  - Longitud de la ciudad.

- **`poi`**  
  - Nombre del punto de interés (POI).

- **`poi_latitude`**  
  - Latitud del POI.

- **`poi_longitude`**  
  - Longitud del POI.

- **`category`**  
  - Categoría principal del POI (ej. cultural, parque, histórico).

- **`subcategory`**  
  - Subcategoría más específica del POI (ej. museo, iglesia, fuente).

- **`value_en`**  
  - Descripción breve en inglés del POI.

- **`opening_hours`**  
  - Horario de apertura del POI (cuando está disponible).

- **`opening_hours_source`**  
  - Fuente o método utilizado para obtener el horario de apertura.

- **`rating`**  
  - Valoración media del POI en una escala de 0 a 5.

- **`reviews_count`**  
  - Número total de reseñas asociadas al POI.

- **`price_level`**  
  - Nivel de precio del POI (generalmente en escala discreta).

- **`google_place_id`**  
  - Identificador obtenido de la API de Google Places durante el proceso de enriquecimiento.  
  - **Prescindible** para el desarrollo del modelo.

- **`match_confidence`**  
  - Nivel de confianza en la correspondencia del POI utilizando datos de OpenStreetMap (OSM) durante el enriquecimiento.

En términos de calidad de datos, se han identificado columnas prescindibles y otras con un alto porcentaje de valores faltantes, lo que hace necesario un proceso de limpieza y selección de variables en la siguiente fase del proyecto.

En conjunto, este análisis ha permitido sentar las bases para el desarrollo del sistema de recomendación, identificando las variables más relevantes y los posibles problemas de calidad de datos.
